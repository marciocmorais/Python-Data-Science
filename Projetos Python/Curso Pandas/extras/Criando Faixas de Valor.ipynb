```python
import pandas as pd
```

```python
dados = pd.read_csv('dados/aluguel.csv', sep=";")
dados.head(10)
```

```python
classes = [0, 2, 4, 6, 100]
```

```python
quartos = pd.cut(dados['Quartos'], classes)
quartos.value_counts()
```

```python
labels = ['0 a 2 quartos', '3 a 4 quartos', '5 a 6 quartos', '7 quartos ou mais']
```

```python
quartos = pd.cut(dados['Quartos'], classes, labels = labels)
quartos.value_counts()
```

```python
dados['Classe'] = pd.cut(dados['Quartos'], classes, labels = labels,include_lowest = True)
dados.head(10)
```
