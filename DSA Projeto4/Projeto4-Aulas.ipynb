## Análise Multivariada

```python
# Scatter Plot (Gráfico de disperção - Bivariável)
sns.scatterplot(x=df["balance"], y=df["salary"])
plt.title("Scatter Plot Entre Saldo e Salário\n")
plt.show()
```

## Numérico x Categórico

```python
df = df[df['response'] != 57008.65331415293]
```

```python
df.groupby(by = ["response"])["salary"].mean()
```

```python
df.groupby(by = ["response"]).salary.median()
```

```python
plt.figure(figsize = (10, 5))
sns.boxplot(x = df.response, y = df.salary)
plt.title("Salário x Resposta\n")
plt.show()
```

```python
# Agrupar educação por salário e calcular a média
df.groupby(by = "education").salary.mean()
```

```python
# Cria uma variável response_flag como tipo numérico onde response "yes" = 1, "no" = 0
df["response_flag"] = np.where(df.response == "yes", 1, 0)
df.head()
```

```python
# Mapa de Correlação
res1 = df.pivot_table(index ="education", columns = "marital", values = "response_flag", aggfunc = "mean")
sns.heatmap(res1, annot = True, cmap = "RdYlGn")
plt.title("Education x Marital x Response Flag\n")
plt.show()
```
